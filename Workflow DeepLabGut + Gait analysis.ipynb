# DeepLabCut
The terminal code used in this workflow applies to Macbook users - slight differences may be applicable to e.g. Windows users.
also please notice when it is said to run 'XXX code', this is without the ''.

## Installing DeepLabCut
The workflow start out with installing DeepLabCut onto your computer

For Macbooks with M1 chip, this tutorial is recommended: https://www.youtube.com/watch?v=hurWt7m41NY 

## Opening DeepLabCut

To start deeplabcut, open the Terminal, navigate to where the deeplabcut folder is located (if located in documents, the code 'cd Documents' can be used to navigate to documents, and run 'conda activate DEEPLABCUT_M1', followed by 'pythonw -m deeplabcut'.

## DeepLabCut workflow

### Creating / loading project

When the GUI interface has started, then click the top bar called 'Manage Project'. Assuming you are starting a new project, name the project and experimenter, and select the videos to be used. OBS, be aware, once a video is selected, it cannot be deselected without restarting the program, so be careful. The optional attributes should not be ticked. A multi-animal project should be used when using videos with more than one animal in the video frames.
If opening an existing project, browse to the folder of your existign project and select the config file. 
After either of these steps, press 'OK' and your porject will be loaded/created.

OBS: depending on how many parts of the animal you wish to label, the config file should be edited at this point to allow for labelling the points. Press 'Edit config file' and a text editor with the config file will pop up. Here you navigate to 'bodyparts:' and enter the parts you wish to track.
For example:
bodyparts:
- snout1
- snout2
- snout3
- leftfrontpaw
- rightfrontpaw
- leftbackpaw
- rightbackpaw
- torsoleft
- torsomiddle
- torsoright
- tail

When you are finished, just close the text editor window and proceed.

### Extract frames
The next step is 'Extract Frames'. Here, the default options should not be changed, but to ensure we are all on the same page, this is how the options should be ticked:

Then press 'OK' and continue to 'Label Data'


### Label Data
The program should already have selected the config file used in your project, other wise brose and choose it. then press 'Label Frames', and then a new windoes should pop up. In that window, press 'Load frames', and you should be in the folder called 'labeled-data' within your project folder. In here, choose the folder for the video you wish to label. then the frames from the video is loaded, and you right-click on the body parts to mark them. the markers can be adjusted in size, and dragged around for correct placement. You press 'Save' and close the pop-up window.


### Create Training
The model is created during the training session. In this part we specify the details of our model. For this, specify the network you wish to use. We used the resnet_50.
Default settings are used.


### Train
In this part we specify the maximum number of iterations, how often the model should be saved, and to display the number of iterations the model has been through. For the remaining options, default settings are used.

For the fist training session, we set maximum number of iterations to 20.000, but stopped it after 5000, saved the model at every 5000 iterations, and displayed the progres at every 100 iterations.

For the re-training of the model (after the initial predictions) we set maximum number of iterations to 20.000, saved the model at every 1000 iterations, and displayed the progres at every 100 iterations.

Then you press 'OK', and the trainnig commences. This step takes a while, depending on your computer's capacity.

### Evaluate
After the training is finished, we evaluate the model. In the 'Evaluate' tab, you use the default settings and press 'RUN: Evaluate network'

### Analyze videos
Here the model makes predictions about the videos we want to analyze based on the model. We can choose videos that weren't part of the traning phase, and videos that were. Select the videos you want to analyze, and press 'Analyze videos'. Default settings used. This process also takes a while.

### Create videos
In the 'create videos' tab, with the default settings you press 'Build Skeleton', followed by 'RUN'

# Kiehn gait analysis
to install the Kiehn gait analysis tool, open the terminal and run 'git clone https://github.com/raghavian/gait_analysis.git' - This should be done in the base - then rest is done in deeplabcut mode (conda activate DEEPLABCUT_M1).

Preferably, the gait analysis tool should be installed in the same location as DeepLabCut, to avoid additional navigation. To start the gait analysis program, run 'pythonw -m gait_analysis' when in the correct location.

If having problems with starting the program, it might be because there are dependencies lacking. In that case, consult the terminal to see which dependencies are lacking, and pip install them (e.g. pip install pyexifinfo). I had one package I had to install manually without pip install. These problems are solved case-to-case basis, but stackoverflow and google is your friend.
